```python
from sympy import *
import numpy as np

x = Symbol('x')
y = x*math.e**(-x)+1

yprime = y.diff(x)
print(yprime)
```

```python
def f(x):                            #funcion
  return x*math.e**(-x)+1
```

```python
def derf(x):                             #derivada funcion
  return -x*math.e**(-x)+math.e**(-x)
```

```python
n=int(input("numero de iteraciones= "))
xi=float(input("ingrese xi= "))
for i in range(n):
 xi1=xi-(f(xi)/derf(xi))
 xi=xi1
 print("x",i,"=",xi1)             
```
